```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
path_image = "./parrot.jpg" #Read image in grayscale and return in a numpy array
img = cv2.imread(path_image)
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
plt.figure(figsize=(16,9))

plt.subplot(221)
imgplot = plt.imshow(img)
plt.colorbar(orientation='horizontal', ticks=[img.min(), img.max()])

plt.subplot(222)
imgplot = plt.imshow(img[:,:,0], cmap = "gray")
plt.colorbar(orientation='horizontal', ticks=[img.min(), img.max()])
plt.subplot(223)
imgplot = plt.imshow(img[:,:,1],cmap = "gray")
plt.colorbar(orientation='horizontal', ticks=[img.min(), img.max()])

plt.subplot(224)
imgplot = plt.imshow(img[:,:,2], cmap = "gray")

plt.colorbar(orientation='horizontal', ticks=[img.min(), img.max()])
```
