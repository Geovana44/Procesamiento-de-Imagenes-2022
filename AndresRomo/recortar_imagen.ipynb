```python
from cv2 import imread
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = imread('./Numpy_wolf_crop.jpeg', 0) #Read image in grayscale and return in a numpy array
print('Original Image shape = ',img.shape)
print('Maximun pixel value (white) = ',img.max())
fig = plt.figure(figsize=(10,10))
plt.imshow(img, cmap='gray')
plt.colorbar(orientation='horizontal', ticks=[img.min(), img.max()])
```

```python
def Recortar_blancos(image):
  bordes = image == 255
  filas_a_recortar = np.all(bordes, axis=1) == False
  columnas_a_recortar = np.all(bordes, axis = 0) == False 
  image = image[filas_a_recortar,:]
  image = image[:,columnas_a_recortar]
  return image
```

```python
img_recortada = Recortar_blancos(img)
fig = plt.figure(figsize=(10,10))
plt.imshow(img_recortada, cmap='gray')
plt.colorbar(orientation='horizontal', ticks=[img.min(), img.max()])
```
